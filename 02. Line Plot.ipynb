- Take a graph paper! (import matplotlib)

```python
import matplotlib.pyplot as plt
```

- Specify data to plot

| x || 1 | 2 | 3 |

| y || 2 | 4 | 1 |

```python
x = [1,2,3]
y = [2,4,1]
```

```python
# Specify X and Y axis labels
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# Give title to your graph!
plt.title("My first graph!")

# Plot points on graph paper!
plt.plot(x,y)

# Show the plot
plt.show()
```
