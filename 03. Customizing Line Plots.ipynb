# Customizing Line Plots

```python
import matplotlib.pyplot as plt
```

```python
plt.style.available
```

```python
plt.style.use('fivethirtyeight')
```

```python
x = [2,2,3,4,5,6]
y = [2,4,3,5,2,6]
```

```python
import numpy as np
```

```python
plt.plot(x, y, color='red', linewidth=2, linestyle='-.',
        marker='o', markersize=10, markerfacecolor='blue')

plt.xlim(0,7)
plt.ylim(0,7)

plt.xticks(np.arange(0,7,0.5))

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.title("My cool graph")

plt.show()
```
