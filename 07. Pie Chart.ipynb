# Pie Chart

![](https://www.patternfly.org/pattern-library/data-visualization/pie-chart/img/bottom-legend-pie-chart.png)

```python
import matplotlib.pyplot as plt
```

```python
# defining labels
activities = ['eat', 'sleep', 'work', 'play']
```

```python
# portion covered by each label
slices = [3, 7, 8, 6]
```

```python
# color for each label
colors = ['r', 'm', 'g', 'b']
```

```python
plt.pie(slices, labels=activities, colors=colors, shadow=True,
       explode=(0,0,0.1,0), autopct="%1.1f%%", startangle=90)

plt.legend(loc=(1,1))

plt.show()
```
